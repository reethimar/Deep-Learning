```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
#data = pd.read_csv('https://raw.githubusercontent.com/drazenz/heatmap/master/autos.clean.csv')
data = pd.read_csv('C:/Users/reraveen/Jupyter/Dataset/ORMB_Data_version.csv',na_values ='')
data.dtypes
```

```python
# creating instance of labelencoder
labelencoder = LabelEncoder()
column_list = ['PERSON_TYPE',
 'BIRTH_DT',
 'LANGUAGE_CD',
 'DIVISION',
 'CUSTOMER_SEG_FLG',
 'CUSTOMER_TIER_FLG',
 'PERSON_NAME',
 'PARTY_TYPE',
 'PRICING_ATTACHED_TO',
 'PRICEITEM_CD',
 'PRICE_STATUS',
 'CUST_CL_CD',
 'CIS_DIVISION',
 'ACCOUNT_CATEGORY',
 'CURRENCY_CD',
 'BILL_CYC_CD',
 'PRODUCT_CD']
categorical_df = data[column_list]
print(categorical_df.dtypes)
categorical_df=categorical_df.astype('category')
print(categorical_df.dtypes)
imp_mean = SimpleImputer(missing_values=np.nan, strategy='constant')
imp_mean.fit(categorical_df)
categorical_df =imp_mean.transform(categorical_df)
categorical_df=pd.DataFrame(categorical_df)
categorical_df =categorical_df.apply(LabelEncoder().fit_transform)
categorical_df.columns=column_list
categorical_df
# Assigning numerical values and storing in another column
#encoded_df = labelencoder.fit_transform(categorical_df)
```

```python


corr = categorical_df.corr()
print(corr)
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
